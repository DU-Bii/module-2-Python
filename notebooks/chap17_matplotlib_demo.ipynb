## Chargement du (sous-)module

On charge le sous-module `pyplot` de `matplotlib` sous l'alias `plt` :

```python
import matplotlib.pyplot as plt
```

Pour certaines versions un peu anciennes de Jupyter Lab, la ligne suivante est nécessaire pour afficher les graphiques dans le notebook :

```python
%matplotlib inline
```

## Données

Évolution de la concentration d'un produit dans le sang (exprimé en mg/L) en fonction du temps (exprimé en heure).

| Temps (h) | Concentration (mg/L) |
|:---------:|:--------------------:|
|     1     |          3.5         |
|     2     |          5.8         |
|     3     |          9.1         |
|     4     |         11.8         |
|     6     |         17.5         |
|     7     |         21.3         |
|     9     |         26.8         |

```python
temps = [1, 2, 3, 4, 6, 7, 9]
concentration = [5.5, 7.2, 11.8, 13.6, 19.1, 21.7, 29.4]
```

## Nuage de points

`plt.scatter()`

```python
plt.scatter(temps, concentration)
```

Pour éviter la ligne `<matplotlib.collections.PathCollection at ...`, il suffit d'ajouter un `;` à la fin de la dernière ligne de la cellule :

```python
plt.scatter(temps, concentration);
```

On peut ajouter un titre pour le graphique, l'axe des abscisses et l'axe des ordonnées

```python
plt.scatter(temps, concentration)
plt.xlabel("Temps (h)")
plt.ylabel("Concentration (mg/L)")
plt.title("Concentration de produit en fonction du temps");
```

On peut enfin, changer la forme, la taille et la couleur des points.

Choix des *markers* : https://matplotlib.org/stable/api/markers_api.html

```python
plt.scatter(temps, concentration, marker="s", s=50, c="orange")
plt.xlabel("Temps (h)")
plt.ylabel("Concentration (mg/L)")
plt.title("Concentration de produit en fonction du temps");
```

## Lignes

`plt.plot()`

```python
plt.plot(temps, concentration);
```

Cette fois, tous les points n'apparaissent plus et sont reliés par une ligne continue.

On peut à nouveau ajouter des titres pour le graphique et les axes, modifier les couleurs et faire apparaître les points :

```python
plt.plot(temps, concentration, color="green", linestyle="--", linewidth=4, 
         marker="o", markersize=12, markerfacecolor="orange")
plt.xlabel("Temps (h)")
plt.ylabel("Concentration (mg/L)")
plt.title("Concentration de produit en fonction du temps");
```

*Cheat sheets* utiles pour Matplotlib : https://github.com/matplotlib/cheatsheets

## Diagramme en bâtons

```python
bases = ["A", "T", "C", "G"]
distribution = [85, 59, 68, 74]
```

`plt.bar()` : diagramme en bâtons verticaux

```python
plt.bar(bases, distribution);
```

On peut ajouter des titres pour le graphique et les axes, et modifier un peu l'esthétique :

```python
plt.bar(bases, distribution, color="orange", width=0.4)
plt.ylabel("Nombre de bases")
plt.xlabel("Bases")
plt.title("Nombre de bases dans ma séquence");
```

`plt.barh()` : diagramme en bâtons horizontaux

```python
plt.barh(bases, distribution, color="palegreen", edgecolor="green", height=0.6, linewidth=2)
plt.ylabel("Bases")
plt.xlabel("Nombre de bases")
plt.title("Nombre de bases dans ma séquence");
plt.savefig("composition_bases.png", dpi=300)
```
